```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("./datasets/simulation_data.csv")
df.info()
```

```python
df.head()
```

```python
df.head(1)
```

```python
df = df.drop(columns=["Class"])
df.info()
```

```python
data_per_file = 10_000
print(f"approx number of files for {data_per_file} data each file: {len(df) / 10_000}")
```

```python
total_files = math.ceil(len(df) / data_per_file)
batched_dfs = np.array_split(df, total_files)

for i, batch in enumerate(batched_dfs):
    print(f"Batch {i+1} size: {len(batch)}")
    batch.to_csv(f"./datasets/simulation_perday_topredict/data_to_predict_{i+1}.csv", index=False)
```
